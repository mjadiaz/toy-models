```python
# calculate the entropy for the split in the dataset
def entropy(class0, class1):
	return -(class0 * np.log2(class0) + class1 * np.log2(class1))
```

```python
def fit_gp(data,z, kernel_type = 'gaussian', h=1):
    kernel = ConstantKernel(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
    model = GaussianProcessRegressor(kernel)
    model.fit(data, z)
    return model
def gp_predict(data,model):
    mu, sigma = model.predict(data, return_std=True)
    return mu, sigma
```
