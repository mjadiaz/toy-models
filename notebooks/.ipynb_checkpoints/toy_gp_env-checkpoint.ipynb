## Defining the observable

```python
def get_simulation_observable(x1,x2, observable_value, sigma):
    '''
    Receives the parameter space point and return the state
    
    Args:
    ----
    x1
    x2
    
    Returns:
    -------
    state = np.array([obs])
    '''
    model = lambda x1, x2: (2 + np.cos(x1/2)*np.cos(x2/2))**5
    lh_real = np.exp(-(model(x1,x2) - observable_value)**2/(2*sigma**2))
    return lh_real, model(x1,x2)
def get_simulation_observable_actual(x1,x2, observable_value, sigma):
    '''
    Receives the parameter space point and return the state
    
    Args:
    ----
    x1
    x2
    
    Returns:
    -------
    state = np.array([obs])
    '''
    model = lambda x1, x2: (2 + np.cos(x1/2)*np.cos(x2/2))**5
    real = model(x1,x2)
    norm = np.log(real)
    return real, norm
```

```python
def distance(variable, goal, stability=1e-5,acceptance=2):
    return -np.log(((variable-goal)**2+stability)*(acceptance)**(-2))
```

```python
bottom, top = 10, 15
samples = 100
x = np.linspace(bottom, top , samples)
y = np.linspace(bottom, top , samples)

X,Y = np.meshgrid(x,y)

X_ = np.array([minmax(i, [bottom, top ], [-1,1])[0] for i in X.flatten()])
Y_ = np.array([minmax(i, [bottom, top ], [-1,1])[0] for i in Y.flatten()])

lh,z = get_simulation_observable(X,Y,100,10)
d = distance(z,100, stability=20**2*np.exp(-1),acceptance=20)
```

```python
fig, ax = plt.subplots(2,2, figsize=(16,16))
cb = ax[0,0].scatter(X,Y,c=z.flatten())
ax[0,0].title.set_text('EggBox model')
fig.colorbar(cb, ax=ax[0,0])

cb = ax[0,1].scatter(X_,Y_,c=lh.flatten())
ax[0,1].title.set_text('Likelihood')
fig.colorbar(cb, ax=ax[0,1])

cb = ax[1,1].scatter(X_,Y_,c=d.flatten())
ax[1,1].title.set_text('Log Distance (Action and Physics Value Space)')
fig.colorbar(cb, ax=ax[1,1])

cb = ax[1,0].scatter(X_,Y_,c=np.log(z).flatten())
fig.colorbar(cb, ax=ax[1,0])
ax[1,0].title.set_text('Normalized parameters - Log EggBox (State Space)')
plt.show()
```
