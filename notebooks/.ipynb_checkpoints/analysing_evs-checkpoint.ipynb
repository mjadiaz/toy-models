```python
import toy_models
import matplotlib.pyplot as plt
import gym
import warnings
import numpy as np
warnings.filterwarnings("ignore")
```

```python
env_name = 'ToyFunction2d-v1'
env = gym.make(env_name)
```

```python
x,y,X,Y = env.simulator.generate_space()
lh = env.simulator.run(x,y)
plt.scatter(x,y,c=lh)
plt.colorbar()
plt.show
```

```python
rewards = env.get_reward(np.vstack((x,y)).T)
plt.scatter(x,y,c=rewards)
plt.colorbar()
```
