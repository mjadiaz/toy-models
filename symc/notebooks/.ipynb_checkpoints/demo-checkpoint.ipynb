### Rosenbrock function üçå

```python
sync_config = SYNC_CONFIG
sync_config['simulation'] = 'rosenbrock2d'
sync_config['bounds'] = [[-1.5, 1.5],[0.,2.]]
pprint(dict(sync_config))
```

```python
sim = SyncSimulation()
```

#### Random uniform sampling

```python
pprint(sim.sample(3))
```

```python
x = sim.sample(10000)
y = sim(x)
```

```python
labels = [r'$\theta_1$', r'$\theta_2$']
ax = color_plot(x, y, labels, cmap='Blues_r')
norm = plt.Normalize(y.min(), y.max())
sm = plt.cm.ScalarMappable(cmap="Blues_r", norm=norm)
sm.set_array([])
# Remove the legend and add a colorbar
#ax.get_legend().remove()
ax.figure.colorbar(sm, orientation='horizontal')

plt.show()
```
